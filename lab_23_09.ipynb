```python
# Step 1: Import required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)
```

```python
# Check the data types of each feature
data_types = titanic_df.dtypes
print(data_types)
```

```python
# Generate summary statistics
summary_stats = titanic_df.describe()
print(summary_stats)
```

```python
# List the names of columns/features in the dataset
column_names = titanic_df.columns.tolist()
print(column_names)
```

```python
# Plotting numeric features
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Check survival count
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.show()
```

```python
# Gender-wise survival rate
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Class-wise survival rate
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Siblings/Spouses Count
sns.barplot(x='SibSp', y='Survived', data=titanic_df)
plt.title('Survival Rate by Number of Siblings/Spouses')
plt.show()
```

```python
# Overall age distribution
plt.figure(figsize=(10, 5))
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Overall Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Class-wise age distribution
plt.figure(figsize=(10, 5))
sns.boxplot(x='Pclass', y='Age', data=titanic_df)
plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
# Impute missing Age values with median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
```

```python
# Drop unnecessary features
titanic_df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
# Recode categorical features
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
titanic_df['Embarked'] = titanic_df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```
